```python
import tensorflow as tf
tf.__version__
# List all physical devices
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        # Enable memory growth for each GPU
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        # Memory growth must be set before any TensorFlow operations
        print(e)
```

```python
trainPath = 'New_Dataset/train_directory'
testPath = 'New_Dataset/test_directory'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,zoom_range=0.2,shear_range=0.2)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train = train_datagen.flow_from_directory(trainPath,target_size=(140,140),batch_size=16)
test = test_datagen.flow_from_directory(testPath,target_size=(140,140),batch_size=16)
```

```python
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Model
```

```python
Xception = Xception(include_top=False,input_shape=(140,140,3))
```

```python
x = Flatten()(Xception.output)
```

```python
output = Dense(4,activation='softmax')(x)
```

```python
Xception = Model(Xception.input,output)
```

```python
Xception.summary()
```

```python
Xception.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
Xception.fit(train,validation_data=test,epochs=15, batch_size = 16)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = Xception.evaluate(test)

# Print the final test accuracy
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
Xception.save("trained.keras")
```
